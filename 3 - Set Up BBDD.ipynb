#     5. Conversion de tipo de datos de la columna AvgTalkDuration


```python
bbdd['AVGTALKDURATION'].replace(' ', '00:00:00', inplace=True)
```

```python
bbdd['AVGTALKDURATION'].info()
```

```python
bbdd['AVGTALKDURATION'] = bbdd['AVGTALKDURATION'].astype(str)
```

```python

bbdd['AVGTALKDURATION'] = pd.to_timedelta(bbdd['AVGTALKDURATION'])
 #Convierte la duración a minutos
bbdd['AVGTALKDURATION'] = bbdd['AVGTALKDURATION'].dt.total_seconds() / 60
```

```python
bbdd['AVGTALKDURATION'].info()
```

```python
bbdd['AVGTALKDURATION'] = bbdd['AVGTALKDURATION'].astype(str).apply(lambda x: x[10:] if 'days' in x else x)

print(bbdd)
```

```python
bbdd
```

```python
bbdd['AVGTALKDURATION'] = pd.to_datetime(bbdd['AVGTALKDURATION'], unit='m').dt.strftime('%H:%M:%S')
```

```python
bbdd['AVGTALKDURATION'] = pd.to_datetime(bbdd['AVGTALKDURATION'], format='%H:%M:%S').dt.strftime('%H:%M:%S')
```
