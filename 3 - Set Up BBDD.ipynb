#     6. Conversion de tipo de datos de la columna AvgTalkDuration
####     En primer lugar completamos los llamados no atendido con la duración de llamada 0:00:00 dado que no fueron realizados efectivamente

```python
bbdd['AVGTALKDURATION'].replace(' ', '00:00:00', inplace=True)
```

####   Convertimos el tipo de dato en datetime con  formado de horas, minutos y segundos para que nos permita trabajar de una mejor manera en el futuro y aplicamos un lambda en para modificar el tipo de dato a timedelta que sera mas util para los proximos calculos

```python
import pandas as pd
from datetime import datetime, timedelta

bbdd['AVGTALKDURATION'] = pd.to_datetime(bbdd['AVGTALKDURATION'], format='%H:%M:%S').dt.time


bbdd['AVGTALKDURATION'] = bbdd['AVGTALKDURATION'].apply(lambda x: timedelta(hours=x.hour, minutes=x.minute, seconds=x.second))

bbdd['AVGTALKDURATION_S'] = bbdd['AVGTALKDURATION'].dt.total_seconds()
```
